### Cenário 1

A empresa Carri Construtora contratou a empresa que você trabalha para encontrar possíveis compradores para seus novos empreendimento, disponíveis pelo programa Minha casa, Minha vida. 
A empresa quer entender as necessidades dos clientes e quer informações como:
- Quais clientes devemos abordar?
- Esse cliente está em busca de investir em um imóvel ou comprar para moradia?
- Qual é a quantidade ideal de quartos para os imóveis com base no número de filhos dos clientes?

Se um cliente tem 0-1 filhos, oferecer imóveis com 2 quartos.
Se um cliente tem 2-3 filhos, oferecer imóveis com 3 quartos.
Se um cliente tem 4 ou mais filhos, oferecer imóveis com 4 quartos ou mais.


**Se renda anual < 104k, então pode participar do programa Minha casa, Minha vida**

```python
# Salvando df apenas com dados que podem ser úteis para esse cenário 
drop_construtora = {
    'Restrição Alimentar',
    'Streaming',
    'Uso diário protetor solar',
    'Compra e Lê Livros Todos os Anos',
    'Quantidade de Livros Comprados',
    'Autor Favorito'}

df_construtora = df.drop(columns= drop_construtora)
df_construtora.head()
```

```python
# Verificando Coluna de Renda Anual

df_construtora['Renda Anual'].describe([])
```

```python
# Como a Renda Anual máxima é abaixo de 104k, todos podem participar do Minha casa, Minha vida
df_construtora['Minha casa, Minha vida'] = True
clientes_abordar = df_construtora[df_construtora['Minha casa, Minha vida'] == True]

print('Lista dos clientes que devem ser abordados pela construtora:')
print(clientes_abordar['Nome'])
```

```python
# Quem já tem casa própria (financiamento ou quitada) está mais voltado para imóvel como investimento
# Do contrário, são imóveis para moradia

df_construtora['Finalidade do Imóvel'] = np.where((df_construtora.Moradia == 'Quitada')|(df_construtora.Moradia == 'Financiamento'), 'Investimento', 'Moradia')
```

```python
# Esse cliente está em busca de investir em um imóvel ou comprar para moradia?

print('Lista dos clientes e finalidade do imóvel:')
print(df_construtora[['Nome', 'Finalidade do Imóvel']])
```

```python
# Definindo condições e valores correspondentes para o número de quartos 
condicoes = [
    (df_construtora['Número de Filhos'] <= 1),
    (df_construtora['Número de Filhos'] <= 3),
    (df_construtora['Número de Filhos'] >= 4)
]

valores = ['2', '3', '4 ou mais']

# Criando a nova coluna 'Número de Quartos'
df_construtora['Número de Quartos'] = np.select(condicoes, valores, default='Desconhecido')
```

```python
# Qual a quantidade ideal de quartos para ofertar?

print('Lista dos clientes, finalidade do imóvel e número de quartos recomendados:\n')
print(df_construtora[['Nome', 'Finalidade do Imóvel', 'Número de Quartos']])
```

### Cenário 3

Uma influenciadora digital de bem-estar gostaria de analisar possíveis empreendimentos dentro das diferentes propostas que recebe. Ela está interessada em entender como as características demográficas e comportamentais de seu público podem influenciar as decisões sobre quais parcerias e projetos realizar.

Seu objetivo é identificar padrões e insights que a ajudem a decidir quais propostas são mais adequadas para seu público-alvo, considerando aspectos como a faixa etária, hábitos de consumo, atividades físicas, leitura, e questões de saúde como intolerância ao glúten e lactose.

- Qual é a idade média dos seguidores?
- Existe uma correlação entre a idade dos seguidores e o número de filhos?
- Quantos seguidores indicaram que fazem uso diário de protetor solar?
- Qual a porcentagem de seguidores que frequentam academia regularmente?
- Quantos seguidores têm restrições alimentares (glúten e/ou lactose)?
- Existe uma relação entre a frequência de ida à praia e o uso diário de protetor solar (se forem imputados os valores nulos)?
- Quantos seguidores indicaram que compram e leem livros anualmente?
- Qual a média de livros comprados por ano entre aqueles que indicaram leitura anual?


```python
# Salvando df apenas com dados que podem ser úteis para esse cenário 
drop_influencer = {
    'Telefone', 
    'Renda Anual', 
    'Moradia',
    'Streaming'}

df_influencer = df.drop(columns= drop_influencer)
df_influencer.head()
```

```python
# Separando as restrições 
restricao = df_influencer['Restrição Alimentar'].str.get_dummies(sep = ' e ').drop(columns = ['Nenhuma'])

df_influencer = pd.concat([df_influencer, restricao], axis = 1)
```

```python
# Excluindo a coluna de Restrições Alimentares

df_influencer = df_influencer.drop(columns= 'Restrição Alimentar')
```

```python
df_influencer.describe(include='all')
```

```python
# Qual é a idade média dos seguidores?
df_influencer['Idade'].mean()
```

```python
# Existe uma correlação entre a idade dos seguidores e o número de filhos?
correlacao = df_influencer['Idade'].corr(df_influencer['Número de Filhos'])

print(correlacao)
print('Correlação Positiva Moderada: conforme a idade dos seguidores aumenta, o número de filhos também tende a aumentar, mas essa relação não é muito forte.')
```

```python
# Quantos seguidores indicaram que fazem uso diário de protetor solar?
df_influencer['Uso diário protetor solar'].value_counts().get('Sim', 0)

```

```python
# Qual a porcentagem de seguidores que frequentam academia regularmente?
frequenta_academia = df_influencer['Faz Academia'].value_counts(normalize=True).get('Sim', 0) * 100
print(f'Porcentagem de seguidores que frequentam academia regularmente: {frequenta_academia:.2f}%')
```

```python
# Quantos seguidores têm restrições alimentares (glúten e/ou lactose)?
restricao_gluten = df_influencer[df_influencer['Glúten'] == 1].shape[0]
restricao_lactose = df_influencer[df_influencer['Lactose'] == 1].shape[0]
restricao_ambas = df_influencer[(df_influencer['Glúten'] == 1) & (df_influencer['Lactose'] == 1)].shape[0]
total_restricoes = restricao_gluten + restricao_lactose - restricao_ambas

print(f"Número de seguidores com restrição a glúten: {restricao_gluten}")
print(f"Número de seguidores com restrição a lactose: {restricao_lactose}")
print(f"Número de seguidores com ambas as restrições (glúten e lactose): {restricao_ambas}")
print(f"Número total de seguidores com pelo menos uma restrição alimentar (glúten e/ou lactose): {total_restricoes}")

```

```python
# Existe uma relação entre a frequência de ida à praia e o uso diário de protetor solar (se forem imputados os valores nulos)?

df_influencer['Vai na Praia Mensalmente'].fillna(df_influencer['Vai na Praia Mensalmente'].mode()[0], inplace=True)
df_influencer['Uso diário protetor solar'].fillna(df_influencer['Uso diário protetor solar'].mode()[0], inplace=True)

relacao_praia_protetor = pd.crosstab(df_influencer['Vai na Praia Mensalmente'], df_influencer['Uso diário protetor solar'])

print("Relação entre Frequência de Ida à Praia e Uso Diário de Protetor Solar:")
print(relacao_praia_protetor.to_string())
```

```python
# Cálculo das proporções
total_nao_praia = relacao_praia_protetor.loc['Não'].sum()
total_sim_praia = relacao_praia_protetor.loc['Sim'].sum()

proporcao_nao_praia_sim = relacao_praia_protetor.loc['Não', 'Sim'] / total_nao_praia
proporcao_sim_praia_sim = relacao_praia_protetor.loc['Sim', 'Sim'] / total_sim_praia

print("\nProporção de Uso Diário de Protetor Solar:")
print(f"Para quem não vai à praia mensalmente: {proporcao_nao_praia_sim:.2%}")
print(f"Para quem vai à praia mensalmente: {proporcao_sim_praia_sim:.2%}")
```
