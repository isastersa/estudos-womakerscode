```python
import pandas as pd
import numpy as np
```

```python
df_saude_sono = pd.read_csv('saude_do_sono_estilo_vida.csv')
df_saude_sono.columns
```

2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mode]

```python
df_saude_sono.groupby('Profissão')['Duração do sono'].agg(['mean', np.median, pd.Series.mode])
```

3. Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?

```python
total_eng_soft = df_saude_sono[df_saude_sono['Profissão'] == 'Eng. de Software']['Identificador'].count()
is_obeso = df_saude_sono['Categoria IMC'] == 'Obesidade'
is_eng_soft = df_saude_sono['Profissão'] == 'Eng. de Software'
obesos_eng_soft = df_saude_sono[is_eng_soft & is_obeso]['Identificador'].count()

porc_obesos_eng_soft = obesos_eng_soft/total_eng_soft
print(porc_obesos_eng_soft)
```

5. Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono? (Use o método ‘isin’, considere a média)

```python
media_sono_enf = df_saude_sono[df_saude_sono['Profissão'] == 'Enfermeiro(a)']['Duração do sono'].mean()
media_sono_med = df_saude_sono[df_saude_sono['Profissão'] == 'Médico(a)']['Duração do sono'].mean()

print(f'Média de duração do sono do Enfermeiro(a): {media_sono_enf:.2f}')
print(f'Média de duração do sono do Médico(a): {media_sono_med:.2f}')
```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
df_acima_70 = df_saude_sono.loc[df_saude_sono['Frequência cardíaca'] > 70]
df_70_ou_menos = df_saude_sono.loc[df_saude_sono['Frequência cardíaca'] <= 70]
```

```python
media_passos_acima_70 = df_acima_70['Passos diários'].mean()
media_passos_70_ou_menos = df_70_ou_menos['Passos diários'].mean()

print(f'Pessoas com frequência cardíaca menor ou igual a 70: {media_passos_70_ou_menos:.2f}')
print(f'Pessoas com frequência cardíaca maior que 70: {media_passos_acima_70:.2f}')
```
